```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

def forecast_statsmodels(data, forecast_periods=10):
    model = sm.tsa.ARIMA(data, order=(1, 1, 1))  # Ajusta el modelo ARIMA según tu serie temporal
    results = model.fit()

    forecast = results.get_forecast(steps=forecast_periods)
    forecast_index = pd.date_range(start=data.index[-1], periods=forecast_periods + 1, freq='M')[1:]
    
    forecast_mean = forecast.predicted_mean
    forecast_ci = forecast.conf_int()

    plt.figure(figsize=(10, 6))
    plt.plot(data.index, data, label='Actual')
    plt.plot(forecast_index, forecast_mean, color='red', label='Forecast')
    plt.fill_between(forecast_index, forecast_ci.iloc[:, 0], forecast_ci.iloc[:, 1], color='red', alpha=0.2)
    plt.legend()
    plt.show()

# Uso de ejemplo:
# Supongamos que 'tu_data' es tu serie temporal (puedes cargarla desde un archivo CSV o DataFrame)
# forecast_statsmodels(tu_data)
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Cargar datos desde el archivo CSV
file_path = 'supplemental_clinical_data.csv'
data = pd.read_csv(file_path)

# Asumiendo que estás trabajando con una serie temporal específica, por ejemplo, 'updrs_1'
serie_temporal = data['updrs_1'].dropna()

# Pronóstico con statsmodels
def forecast_statsmodels(data, forecast_periods=10):
    model = sm.tsa.ARIMA(data, order=(1, 1, 1))
    results = model.fit()

    forecast = results.get_forecast(steps=forecast_periods)
    forecast_index = pd.date_range(start=data.index[-1], periods=forecast_periods + 1, freq='M')[1:]
    
    forecast_mean = forecast.predicted_mean
    forecast_ci = forecast.conf_int()

    plt.figure(figsize=(10, 6))
    plt.plot(data.index, data, label='Actual')
    plt.plot(forecast_index, forecast_mean, color='red', label='Forecast')
    plt.fill_between(forecast_index, forecast_ci.iloc[:, 0], forecast_ci.iloc[:, 1], color='red', alpha=0.2)
    plt.legend()
    plt.show()
```
