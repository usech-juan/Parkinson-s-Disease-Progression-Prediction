```python
import pandas as pd
import matplotlib.pyplot as plt  


supplemental_data = pd.read_csv("supplemental_clinical_data.csv")
#print("Raw data: ", supplemental_data.count())
print("MISSING VALUES: ", supplemental_data.isnull().sum())

print(supplemental_data["updrs_1"].mean())
print(supplemental_data["updrs_2"].mean())
print(supplemental_data["updrs_3"].mean())
print(supplemental_data["updrs_4"].mean())
print(supplemental_data["updrs_1"].mode())
print(supplemental_data["updrs_2"].mode())
print(supplemental_data["updrs_3"].mode())
print(supplemental_data["updrs_4"].mode())
supplemental_data.columns
supplemental_data
```

```python
visit_month = supplemental_data["visit_month"].unique()
visit_month
```

```python
mean_updrs_1 = supplemental_data["updrs_1"].mean()
mean_updrs_2 = supplemental_data["updrs_2"].mean()
mean_updrs_3 = supplemental_data["updrs_3"].mean()
mode_updrs_4 = supplemental_data["updrs_4"].mode()[0]
supplemental_data["updrs_1"].fillna(mean_updrs_1, inplace = True)
supplemental_data["updrs_2"].fillna(mean_updrs_2, inplace = True)
supplemental_data["updrs_3"].fillna(mean_updrs_3, inplace = True)
supplemental_data["updrs_4"].fillna(mean_updrs_4, inplace = True) #Used due to the value 0 has 999 repetitions so a mean would do nothing
#supplemental_data["upd23b_clinical_state_on_medication"].value_counts() ##Used to know if it was a "Two sides response" (Hypothesis confirmed)
supplemental_data["upd23b_clinical_state_on_medication"].fillna("Off", inplace=True) #We assume that if they're not directly registered as "On" the value will be Off due to how internal medical processes tend to be managed
supplemental_data.isnull().sum()
```

```python
visit_month = supplemental_data["visit_month"]
updrs_1 = supplemental_data["updrs_1"]
updrs_2= supplemental_data["updrs_2"]
updrs_3 = supplemental_data["updrs_3"]
updrs_4 = supplemental_data["updrs_4"]

plt.scatter(visit_month, updrs_1)
plt.scatter(visit_month, updrs_2)
plt.scatter(visit_month, updrs_3)
plt.scatter(visit_month, updrs_4)
plt.show()

```
